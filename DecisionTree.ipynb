#Working with Decision Tree module in SciKit Learn

To learn about Decition tree go to the link below. http://scikit-learn.org/stable/modules/tree.html.

```python
import pandas as pd
from sklearn import tree
import numpy as np
```

```python
train_url = "train.csv"
train = pd.read_csv(train_url)
train.head()
```

```python
test_url = "test.csv"
test = pd.read_csv(test_url)
test.head()
```

```python
train.describe()
```

```python
train["Age"] = train["Age"].fillna(train["Age"].mean())
```

```python
train.describe()
```

```python
test.describe()
```

```python
test["Age"] = test["Age"].fillna(test["Age"].mean())
test["Fare"] = test["Fare"].fillna(test["Fare"].mean())
```

```python
test.describe()
```

Now let's make some new features based on the data. By now possible list of features are Pclass, Age, SibSp, Parch and Fare. We can see that the Sex of the person can have a huge impact on the survival lets add it to feature list. Also the embarked param can also have impact. So add it too.

```python
train.loc[train["Sex"] == "male", "Sex"] = 0
train.loc[train["Sex"] == "female", "Sex"] = 1

test.loc[test["Sex"] == "male", "Sex"] = 0
test.loc[test["Sex"] == "female", "Sex"] = 1
```

```python
train.describe()
```

```python
train.loc[train["Embarked"] == "S", "Embarked"] = 0
train.loc[train["Embarked"] == "C", "Embarked"] = 1
train.loc[train["Embarked"] == "Q", "Embarked"] = 2

test.loc[test["Embarked"] == "S", "Embarked"] = 0
test.loc[test["Embarked"] == "C", "Embarked"] = 1
test.loc[test["Embarked"] == "Q", "Embarked"] = 2

train["Embarked"] = train["Embarked"].fillna(0)
```

```python
train.describe()
```

Now make a decision tree with everything that is set by default.

```python
decision_tree = tree.DecisionTreeClassifier()
```

Next step is to defining what is our features and what is target that we want to classify.

```python
features = train[["Pclass","Age", "SibSp", "Parch", "Fare", "Sex", "Embarked"]].values
target = train[["Survived"]].values
```

Now call the $\boxed{.fit}$ function on the decision tree to classify Survival from the given features.

```python
decision_tree = decision_tree.fit(features, target)
```

```python
print("The score of the fitting is {}".format(decision_tree.score(features, target)))

feature_keys = np.array(["Pclass","Age", "SibSp", "Parch", "Fare", "Sex", "Embarked"])
feature_importance = np.array(decision_tree.feature_importances_)
importance = pd.DataFrame(feature_importance,feature_keys, columns=["Importance"])

importance
```

Now that desision tree is made, let's predict values from test data, first make test features vector

```python
features_t = test[["Pclass","Age", "SibSp", "Parch", "Fare", "Sex", "Embarked"]].values
```

Now run prediction

```python
prediction = decision_tree.predict(features_t)
```

```python
result = pd.DataFrame(prediction, np.array(test["PassengerId"]), columns = ["Survived"])
```
